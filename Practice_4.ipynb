## Column Names (переведено Google)

- id - уникальный идентификатор для каждого проданного дома.
- date - Дата продажи дома.
- price - Цена каждого проданного дома
- bedrooms - Количество спален
- bathrooms - количество ванных комнат, где 0,5 соответствует комнате с туалетом, но без душа.
- sqft_living - Площадь внутренней жилой площади квартиры в квадратных футах
- sqft_lot - Площадь земельного участка в квадратных футах.
- floors - Количество этажей
- waterfront - Фиктивная переменная, определяющая, выходила ли квартира на набережную или нет.
- view - Индекс от 0 до 4, показывающий, насколько хорош вид на недвижимость.
- condition - Индекс от 1 до 5 о состоянии квартиры
- grade -  Индекс от 1 до 13, где 1-3 соответствует строительству и проектированию зданий, 7 - среднему уровню строительства и проектирования, 11-13 - качественному уровню строительства и проектирования.
- sqft_above - Площадь внутреннего жилого помещения, находящегося над уровнем земли, в квадратных футах.
- sqft_basement - Площадь внутреннего жилого помещения, находящегося ниже уровня земли, в квадратных футах.
- yr_built - Год, когда дом был построен.
- yr_renovated - Год последней реконструкции дома.
- zipcode - В каком районе с почтовым индексом находится дом.
- lat - Широта
- long - Долгота
- sqft_living15 - Площадь внутренней жилой площади дома для ближайших 15 соседей.
- sqft_lot15 - Площадь земельных участков ближайших 15 соседей

## 3 задача

- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
plt.figure(figsize=(12, 9))

matrix = round(df.drop(
        columns=['id', 'date', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
        ).corr(),2)
# matrix[abs(corrMatrix) < 0.1] = 0

sns.heatmap(matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title("Матрица корреляции признаков");
```

```python

price_yr = df.groupby('yr_built')['price'].mean().reset_index()/1000
plt.figure(figsize=(12, 9))
plt.scatter(price_yr['yr_built'], price_yr['price'])
plt.title('Распределение средних цен в зависимости от года постройки')
plt.xlabel('Год постройки дома')
plt.ylabel('Цена, тыс.')
plt.grid(True)
```

```python
plt.figure(figsize=(12, 9))
sns.boxplot(x=df['grade'], y=df['price']/1000, whis=1.0)
plt.xlabel('Класс дома')
plt.ylabel('Цена, тыс.')
plt.title('Распределение цен в зависимости от класса дома')
```

### Выводы:

- с увеличением уровня класса растет разбег по цене;
- для домов с 7-11 классом наблюдается максимальное количество выбросов.
